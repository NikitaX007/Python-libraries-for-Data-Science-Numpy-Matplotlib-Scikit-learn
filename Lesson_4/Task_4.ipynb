Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3

```python
# Создание массива x с помощью функции linspace
x = np.linspace(-3, 3, 51)

# Создание массивов y1, y2, y3, y4 по заданным формулам
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

# Создание объекта Figure и массива объектов Axes
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# Установка зазоров между графиками
fig.subplots_adjust(wspace=0.3, hspace=0.3)

# Построение графиков
ax[0, 0].plot(x, y1)
ax[0, 0].set_title('График y1')
ax[0, 0].set_xlim([-5, 5])

ax[0, 1].plot(x, y2)
ax[0, 1].set_title('График y2')

ax[1, 0].plot(x, y3)
ax[1, 0].set_title('График y3')

ax[1, 1].plot(x, y4)
ax[1, 1].set_title('График y4')

# Отображение графиков
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка стиля для графиков
plt.style.use('fivethirtyeight')

# Загрузка датасета
df = pd.read_csv('/content/creditcard.csv')

# Подсчет количества наблюдений для каждого значения целевой переменной Class
class_counts = df['Class'].value_counts()

# Построение столбчатой диаграммы
class_counts.plot(kind='bar')
plt.title('Распределение классов транзакций')
plt.xlabel('Class')
plt.ylabel('Количество наблюдений')
plt.show()

# Построение столбчатой диаграммы с логарифмическим масштабом
class_counts.plot(kind='bar', logy=True)
plt.title('Распределение классов транзакций (логарифмический масштаб)')
plt.xlabel('Class')
plt.ylabel('Логарифм количества наблюдений')
plt.show()

# Построение гистограмм для признака V1
plt.hist(df[df['Class'] == 0]['V1'], bins=20, density=True, alpha=0.5, color='grey', label='Class 0')
plt.hist(df[df['Class'] == 1]['V1'], bins=20, density=True, alpha=0.5, color='red', label='Class 1')
plt.legend()
plt.title('Распределение признака V1')
plt.xlabel('V1')
plt.ylabel('Плотность распределения')
plt.show()
```
